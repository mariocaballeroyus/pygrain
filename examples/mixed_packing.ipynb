## Mixed Particle Packing

```python
import pygrain3d as pg
print(pg.__version__)
```

```python
length = 100.0
packing = pg.packing.create_packing([length, length, length])
```

```python
packing.add_cylinder_particles(
    aspect_ratio=1.0, diameter=20.0, num=20)

packing.add_cylinder_particles(
    aspect_ratio=2.0, diameter=20.0, num=20)

packing.add_spheroid_particles(
    aspect_ratio=2.0, minor_axis=20.0, num=10)

packing.add_sphere_particles(
    radius=10.0, num=20)

print(f"Generating packing with {packing.num_particles} particles "
      f"and {packing.porosity:.2f} porosity.")
```

```python
packing.generate(max_iterations=50000)
```

```python
pg.io.to_csv(packing, "mixed_packing.csv", periodic=True)
```

```python
with pg.mesh.PackingMesh("packing") as mesh:
    mesh.generate(packing, mesh_size=2.0, periodic=False)
    pg.io.to_stl("mixed_packing_mesh.stl")
```
